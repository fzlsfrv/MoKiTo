# Short trajectories

```python
M          = 10

for i in np.arange(0,10,1):
    for r in range(M):
        file_traj_solute = 'final_states/xt_' + str(i) + '_r' + str(r) + '.dcd'
        traj      = md.load(out_trajectories2 + file_traj_solute, top = inp_dir + pdbfile_solute)   
        Nframes   = traj.n_frames 
        df        = np.squeeze(md.compute_distances(traj, [[0,70]], periodic=True))
        plt.plot(np.arange(0,Nframes+1), np.insert( df, 0, R[i]),'k-')
        plt.plot(0, R[i],'bo')

plt.ylim(-np.pi, np.pi)
```
